# <span style="font-family: Arial; font-weight:bold;font-size:0.8em;color:#00b3e5;">4.a. Do charges of people who smoke differ significantly from the people who don't? :

```python
sns.swarmplot(mydata["smoker"],mydata["charges"])
```

* From the above plot we can say that
1. Yes,People who smoke have more medical charges(more than 60000 as well) than those who dont smoke.
2. People who dont smoke have have low charges(frequency higher on lower side) of medical cost when compared to people who smoke.

# <span style="font-family: Arial; font-weight:bold;font-size:0.8em;color:#00b3e5;">4.b. Does bmi of males differ significantly from that of females? :

```python
sns.boxplot(mydata["sex"],mydata["bmi"])
```
